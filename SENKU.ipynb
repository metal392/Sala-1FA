⦁	Desconto em compras
Peça o valor de uma compra:
⦁	≥ 100: 10% de desconto
⦁	≥ 50 e < 100: 5% de desconto
⦁	< 50: sem desconto


```python
valor_compra = float(input("Digite o valor da compra: "))

if valor_compra >= 100:
    desconto = valor_compra * 0.1
    valor_com_desconto = valor_compra - desconto
    print(f"O valor da compra com desconto é: {valor_com_desconto}")
elif valor_compra >= 50 and valor_compra < 100:
    desconto = valor_compra * 0.05
    valor_com_desconto = valor_compra - desconto
    print(f"O valor da compra com desconto é: {valor_com_desconto}")
else:
    print(f"Não há desconton então o valor é: {valor_compra}")
```

⦁	Ano bissexto
Solicite um ano e informe se ele é bissexto ou não.


```python
#entrada

ano = int(input("Digite um ano: "))

#processamento e saída

if (ano %4 == 0 and ano % 100 != 0) or (ano %400 == 0):
    print("O ano é bissexto")
else:
    print("O ano não é bissexto")
```

⦁	Dia da semana
Peça um número de 1 a 7 e mostre o dia correspondente (ex: 1 = domingo).


```python
#entrada

dia = int(input("Digite um número de 1 a 7: "))

#processamento e saída

if dia == 1:
    print("Domingo")
elif dia == 2:
    print("Segunda-feira")
elif dia == 3:
    print("Terça-feira")
elif dia == 4:
    print("Quarta-feira")
elif dia == 5:
    print("Quinta-feira")
elif dia == 6:
    print("Sexta-feira")
elif dia == 7:
    print("Sábado")
else:
    print("Esse é o dia maluco vai se tratar")
```

⦁	Calculadora simples
Peça dois números e uma operação (+, -, *, /). Use if/elif para calcular o resultado

```python
#entrada

numero1 = float(input("Digite o primeiro número: "))
numero2 = float(input("Digite o segundo número: "))
operacao = input("Digite a operação (+, -, *, /): ")

if operacao == "+":
    resultado = numero1 + numero2
    print(f"O resultado da soma é: {resultado}")
elif operacao == "-":
    resultado = numero1 - numero2
    print(f"O resultado da subt é: {resultado}")
elif operacao == "*":
    resultado = numero1 * numero2
    print(f"O resultado da mult é: {resultado}")
elif operacao == "/":
    resultado = numero1 / numero2
    print(f"O resultado da divisão é: {resultado}")
```

⦁	Positivo, negativo ou zero
Peça um número e diga se ele é positivo, negativo ou zero.


```python
#entrada

numero = int(input("Digite um número: "))

#processamento e saída

if numero > 0:
    print("O número é positivo")

elif numero < 0:
    print("O número é negativo")
else:
    print("O número é zero")
```

⦁	Classificação de idade
Receba a idade e classifique:
⦁	Criança (0–12)
⦁	Adolescente (13–17)
⦁	Adulto (18–59)
⦁	Idoso (60+)



```python
#entrada

idade = int(input("Digite sua idade: "))

#processamento e saída

if idade >= 0 and idade <= 12:
    print("TU É UMA CRIANÇA")
elif idade <= 17:
    print("TU É UM ADOLESCENTE")
elif idade <= 59:
    print("Tu é um adulto")
else:
    print("TU É VELHO")
```

Aprovação de aluno
Peça a nota de um aluno e informe:
Aprovado (nota ≥ 7)
Recuperação (5 ≤ nota < 7)
Reprovado (nota < 5)


```python
#entrada

nome = input("Digite o nome do aluno: ")
nota = int(input("Digite a nota do aluno: "))

#processamento e saída

if nota >= 7:
    print (f"O aluno {nome} aprovado com a nota {nota}")
elif nota >= 5: # Corrigido: indentação e uso de 'elif'
    print(f"O aluno {nome} está em recuperação pela nota ser {nota}")
else:
    print(f"O aluno {nome} foi reprovado por estar com a nota {nota}")
```

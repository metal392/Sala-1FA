Questão 2 cauculo desconto

Crie  um programa que leia o valor da compra e aplique o desconto conforme as regras:

até R$100= sem desconto
De R$101 a R$500: 10% de desconto
Acima de R$500: 20% de desconto

```python
#Questão 2

valor_compra = float(input("Digite o valor da compra: "))

if valor_compra <= 100:
    print("Sem desconto. O valor final é R$", valor_compra)
elif valor_compra <= 500: # This implicitly means > 100 and <= 500
    desconto = valor_compra * 0.1
    valor_final = valor_compra - desconto
    print(f"Com 10% de desconto, o valor final é R${valor_final:.2f}")
else: # This implicitly means > 500
    desconto = valor_compra * 0.2
    valor_final = valor_compra - desconto
    print(f"Com 20% de desconto, o valor final é R${valor_final:.2f}")
```

Questão 3 verificação de triangulo

Crie um programa que leia três valores e verifique  se eles podem formar um triangulo
Para isso, a soma  de dois  lados deve ser maior que o terceiro
Mostre uma mensagem informando se é ou não um triangulo valido

```python
lado1 = int(input("Digite o primeiro lado: "))
lado2 = int(input("Digite o segundo lado: "))
lado3 = int(input("Digite o terceiro lado: "))

#processamento/saida

if (lado1 + lado2 > lado3) and (lado1 + lado3 > lado2) and (lado2 + lado3 > lado1):
    print("Os lados formam um triângulo válido.")
    if lado1 == lado2 == lado3:
        print("É um triângulo equilátero.")
    elif lado1 == lado2 or lado1 == lado3 or lado2 == lado3:
        print("É um triângulo isósceles.")
    else:
        print("É um triângulo escaleno.")
else:
    print("Os lados não formam um triângulo válido.")
```

Questão 4 energia jogo

Crie um programa que leia o a enegria de um personagem (de 0 a 100)
e classifique

0= morto
1 a 30: Crítico
31 a 70: Normal
71 a 100: Saudável


```python
#Questão 4

enegia = int(input("Digite a energia do personagem: "))
if enegia == 0:
    print("O personagem está morto.")
elif enegia >= 1 and enegia <= 30:
    print("O personagem está em estado crítico.")
elif enegia >= 31 and enegia <= 70:
    print("O personagem está em estado normal.")
elif enegia >= 71 and enegia <= 100:
    print("O personagem está saudável.")
else:
    print("Energia inválida. Por favor, insira um valor entre 0 e 100.")
```

Questão 6 – Cálculo de salário com bônus

Crie um programa que leia o salário de um funcionário e sua pontuação de desempenho. O programa deve calcular um bônus de acordo com as regras:
Pontuação menor que 5: sem bônus
Pontuação de 5 a 7: bônus de 5% do salário
Pontuação de 8 a 9: bônus de 10% do salário
Pontuação igual a 10: bônus de 15% do salário
Ao final, mostre o valor do bônus e o salário final com o bônus aplicado.Média menor que 5: Reprovado

```python
#questão 6

pontuacao = int(input("Digite sua pontuação: "))
salario = float(input("Digite seu salário: "))

if pontuacao < 5:
    bonus_percentual = 0
elif pontuacao >= 5 and pontuacao <= 7:
    bonus_percentual = 0.05  # 5%
elif pontuacao >= 8 and pontuacao <= 9:
    bonus_percentual = 0.10  # 10%
elif pontuacao == 10:
    bonus_percentual = 0.15  # 15%
else:
    print("Pontuação inválida. Por favor, insira um valor entre 0 e 10.")
    bonus_percentual = 0 # Define bonus as 0 for invalid input to avoid further errors

bonus_valor = salario * bonus_percentual
salario_final = salario + bonus_valor

print(f"Salário Base: R${salario:.2f}")
print(f"Bônus ({bonus_percentual*100:.0f}%): R${bonus_valor:.2f}")
print(f"Salário Final com Bônus: R${salario_final:.2f}")
```

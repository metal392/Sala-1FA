```python
#escrever um programa que o usuario digita um numero e o programa imprime os numeros pares até ele

numero = int(input("Digite um numero: "))

for i in range(numero + 1):
    if i % 2 == 0:
        print(i)
```

```python
#escrever um programa que verifica se a palavra é um palindromo e imprime a palavra

palavra = input("Digite uma palavra: ")

print (palavra[::-1])
if palavra == palavra[::-1]:
    print("A palavra é um palindromo")
else:
    print("A palavra não é um palindromo")
```
